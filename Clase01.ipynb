#   Guía Interactiva: Funciones en Python

¡Bienvenido! En este cuaderno aprenderemos todo sobre las **funciones** en Python.

## ¿Qué es una función?
Una función es un **bloque de código reutilizable** que realiza una tarea específica. En lugar de escribir el mismo código una y otra vez, lo encapsulamos en una función y la llamamos cuando la necesitamos.

Nos ayudan a mantener los siguientes principios:
* **Reusabilidad:** Escribes el código una vez y lo usas en todos lados.
* **Modularidad:** Divides un problema grande en partes pequeñas (funciones) fáciles de manejar.
* **Legibilidad:** El código con funciones bien nombradas se lee casi como texto natural.

---

### 1. Funciones Incorporadas (Built-in)
Python ya viene con varias funciones listas para usar. Seguramente ya conoces algunas:

```python
# Imprimir en consola
print("¡Hola, mundo!")

# Obtener la longitud de una lista o texto
print("Longitud de la palabra:", len("Python"))

# Obtener el valor máximo y mínimo
print("El valor máximo es:", max(10, 50, 5))
print("El valor mínimo es:", min(10, 50, 5))

# Valor absoluto
print("Valor absoluto de -15:", abs(-15))
```

### 2. Creando nuestras propias funciones
Para definir una función nueva en Python se utiliza la palabra reservada `def`, seguida del nombre de la función, paréntesis `()` y dos puntos `:`. Todo el código que pertenezca a la función debe ir **indentado** (con sangría).

```python
# Ejemplo 1: Función sin parámetros
def saludo_basico():
    print('¡Hola amiga!')

# Para ejecutarla, debemos "llamarla"
saludo_basico()

# Ejemplo 2: Función con un parámetro de entrada
def saludo_personalizado(nombre):
    print('¡Hola ' + nombre + '!')

saludo_personalizado('Leandro')
saludo_personalizado('Python')
```

### 3. La sentencia `return`
El uso de `return` permite realizar dos cosas:
1. Salir de la función (todo el código que esté debajo del `return` no se ejecutará).
2. **Devolver uno o varios valores** al código que llamó a la función, para poder guardarlos en variables.

```python
# Función que devuelve un solo valor
def suma_simple(a, b):
    suma = a + b
    return suma

resultado = suma_simple(10, 5)
print("Resultado de la suma:", resultado)


# Función que devuelve múltiples valores
def suma_y_media(a, b, c):
    suma = a + b + c
    media = suma / 3
    return suma, media

total, promedio = suma_y_media(9, 6, 3)
print("Suma total:", total)
print("Promedio:", promedio)
```

### 4. Tipos de Argumentos
Python es muy flexible al momento de pasar datos (argumentos) a una función.

1. **Por posición:** El primer valor va a la primera variable, el segundo a la segunda, etc.
2. **Por nombre:** Indicamos explícitamente el nombre de la variable (`a=3, b=5`).
3. **Por defecto:** Podemos darle un valor inicial a un parámetro por si el usuario no lo ingresa.
4. **Longitud Variable (`*args` y `**kwargs`):** Para cuando no sabemos cuántos datos nos van a pasar.

```python
# --- ARGUMENTOS POR DEFECTO ---
def operacion(a, b, c=0):
    return a + b + c

print("Con 3 argumentos:", operacion(5, 5, 3))  # 13
print("Con 2 argumentos (c usa su valor por defecto 0):", operacion(4, 3))   # 7

# --- LONGITUD VARIABLE (*args) ---
# El asterisco empaqueta todos los argumentos extra en una Tupla.
def sumar_todos(*numeros):
    total = 0
    for n in numeros:
        total += n
    return total

print("\nSuma variable de 4 números:", sumar_todos(1, 3, 5, 4))
print("Suma variable de 7 números:", sumar_todos(6, 4, 10, 20, 4, 6, 7))
```

### 5. Ámbito de las Variables (Scope): Local vs Global
* **Variables Locales:** Las que creas *dentro* de una función. **Solo existen ahí adentro**. Si intentas imprimirlas fuera, Python te dará error.
* **Variables Globales:** Las que creas en el programa principal. Se pueden leer desde adentro, pero es mala práctica modificarlas (salvo que uses la palabra `global`).

```python
variable_global = "Soy global"

def probando_variables():
    variable_local = "Soy local"
    print("Desde adentro:", variable_global)
    print("Desde adentro:", variable_local)

probando_variables()

print("\nDesde afuera:", variable_global)
# Si descomentas la siguiente línea, dará error porque 'variable_local' no existe afuera:
# print(variable_local)
```

### 6. Buenas Prácticas: Documentación y Anotaciones
Una función bien escrita explica qué hace. Para eso usamos los **Docstrings** (triples comillas `"""`) y los **Type hints** (pistas de tipos de datos, indicando con `:` el tipo de entrada y con `->` el tipo de salida).

```python
def calcular_media(muestra: list) -> float:
    """
    Función que calcula la media de una muestra de números.

    Parámetros:
    muestra (list): Es una lista de números enteros o flotantes.

    Devuelve (float): La media aritmética de los números.
    """
    return sum(muestra) / len(muestra)

# Podemos ver la documentación llamando a la función interna help()
help(calcular_media)

print("El promedio es:", calcular_media([2.3, 5.7, 6.8, 9.7, 12.1, 15.6]))
```

### 7. 💻 Ejercicios Prácticos Integradores
A continuación, varios ejemplos de la vida real donde encapsulamos lógica dentro de funciones para poder reutilizarla sin copiar y pegar código.

```python
# Ejercicio A: Calcular Factorial
def factorial(n: int) -> int:
    """Calcula el factorial de un número entero positivo."""
    tmp = 1
    for i in range(n):
        tmp *= (i + 1)
    return tmp

print(f"Factorial de 4 es: {factorial(4)}")
print(f"Factorial de 10 es: {factorial(10)}")
print("-"*30)

# Ejercicio B: Comprobador de Años Bisiestos
def es_bisiesto(fecha: int):
    if fecha % 400 == 0 or (fecha % 100 != 0 and fecha % 4 == 0):
        return f"El año {fecha} es bisiesto."
    else:
        return f"El año {fecha} NO es bisiesto."

print(es_bisiesto(2024))
print(es_bisiesto(2025))
print("-"*30)

# Ejercicio C: Dibujando con bucles anidados en una función
def dibujar_rectangulo(anchura, altura):
    for i in range(altura):
        for j in range(anchura):
            print("* ", end="")
        print()  # Salto de línea al terminar la fila

print("Dibujando Rectángulo 5x3:")
dibujar_rectangulo(5, 3)

print("\nDibujando Rectángulo 2x2:")
dibujar_rectangulo(2, 2)
```
