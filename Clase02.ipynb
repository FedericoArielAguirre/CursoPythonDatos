## 1. Argumentos y Retorno

Una función sin argumentos es rígida, solo hace una cosa específica. Los **argumentos** permiten que la función trabaje con datos distintos cada vez que es llamada. El **retorno** (`return`) permite extraer el resultado para usarlo después.

```python
# ❌ Función rígida (depende de variables globales)
a = 3
b = 5
def escribe_media_rigida():
    media = (a + b) / 2
    print(f"La media es: {media}")

escribe_media_rigida()

# ✅ Función flexible con argumentos y retorno
def calcula_media(x, y):
    resultado = (x + y) / 2
    return resultado

mi_media = calcula_media(10, 20)
print(f"La media calculada y retornada es: {mi_media}")
```

### 1.1. Múltiples valores de retorno
En Python, una función puede devolver más de un valor separándolos por comas. Internamente, Python los empaqueta en una **tupla**.

```python
def operaciones_basicas(a, b):
    suma = a + b
    resta = a - b
    multiplicacion = a * b
    return suma, resta, multiplicacion # Devuelve una tupla

# Podemos desempaquetar el resultado en variables separadas
s, r, m = operaciones_basicas(10, 5)
print(f"Suma: {s}, Resta: {r}, Multiplicación: {m}")
```

### 1.2. Argumentos Variables: `*args` y `**kwargs`
¿Qué pasa si queremos calcular la media de 3, 4 o 100 números?

* `*args`: Permite recibir una cantidad indefinida de argumentos posicionales. Internamente se agrupan en una **tupla**.
* `**kwargs`: Permite recibir argumentos con nombre (`clave=valor`). Internamente se agrupan en un **diccionario**.

⚠️ **Regla de oro para el orden**: Siempre debes poner primero los parámetros fijos, luego `*args`, y al final `**kwargs`.

```python
# Usando *args
def sumar_todos(*args):
    print(f"Datos recibidos (*args): {args} (Tipo: {type(args)})")
    return sum(args)

print("Suma total:", sumar_todos(1, 2, 3, 4, 5))
print("-" * 30)

# Usando **kwargs
def presentar_alumno(**kwargs):
    print(f"Datos recibidos (**kwargs): {kwargs} (Tipo: {type(kwargs)})")
    for clave, valor in kwargs.items():
        print(f"{clave.capitalize()}: {valor}")

presentar_alumno(nombre="Ana", edad=21, lenguaje="Python")
```

## 2. Funciones Lambda (Anónimas)
Las funciones **lambda** son funciones pequeñas, de una sola línea y sin nombre. Se usan mucho para operaciones rápidas o para pasarlas como argumento a otras funciones (como `map` o `filter`).

**Sintaxis:** `lambda argumentos: expresion`

```python
# Función tradicional con def
def cuadrado_def(x):
    return x ** 2

# Equivalente usando lambda
cuadrado_lambda = lambda x: x ** 2

print("Con def:", cuadrado_def(5))
print("Con lambda:", cuadrado_lambda(5))
```

### 2.1. Funciones de Orden Superior: `map()` y `filter()`
* `map(funcion, lista)`: Aplica la función a cada elemento de la lista y devuelve los resultados.
* `filter(funcion, lista)`: Devuelve solo los elementos de la lista donde la función da `True`.


```python
numeros = [1, 2, 3, 4, 5, 6]

# MAP: Elevar al cuadrado
cuadrados_map = list(map(lambda x: x ** 2, numeros))
print("Usando map:", cuadrados_map)

# FILTER: Obtener pares
pares_filter = list(filter(lambda x: x % 2 == 0, numeros))
print("Usando filter:", pares_filter)

print("-" * 30)
# ALTERNATIVA MODERNA: Comprensión de listas
cuadrados_comp = [x ** 2 for x in numeros]
pares_comp = [x for x in numeros if x % 2 == 0]
print("Cuadrados con comprensión:", cuadrados_comp)
print("Pares con comprensión:", pares_comp)
```

## 3. El uso de `assert` (Testing)
El `assert` verifica que una condición sea verdadera (`True`). Si lo es, el programa sigue normal. Si es falsa (`False`), el programa se detiene lanzando un `AssertionError`.

**Sintaxis recomendada**: `assert condicion, "Mensaje de error"`

NUNCA usar paréntesis englobando la condición y el mensaje `assert(False, "error")`. Python lo tomará como una Tupla, y las tuplas llenas siempre dan `True`, por lo que tu `assert` **nunca fallará**.

```python
# Ejemplo correcto de assert
x = 10
assert x == 10, "x debería valer 10"
print("El assert pasó correctamente, todo en orden. ✓")

def es_mayor_de_edad(edad):
    return edad >= 18

# Pruebas automáticas de nuestra función
assert es_mayor_de_edad(20) == True, "Falló para alguien de 20 años"
assert es_mayor_de_edad(15) == False, "Falló para alguien de 15 años"
print("Las pruebas de la función pasaron correctamente. ✓")

# Si descomentas la siguiente línea, el programa se detendrá con error:
# assert es_mayor_de_edad(17) == True, "Esto fallará intencionalmente"
```
