# 🐍 Python: Diccionarios y Excepciones

Conceptos fundamentales para programar aplicaciones reales y robustas en Python:

1. **Diccionarios**: Estructuras para almacenar datos de forma organizada (pares clave-valor).
2. **Excepciones**: Cómo evitar que nuestro programa colapse cuando ocurre un error.
3. **Proyecto Integrador**: Un sistema de registro de compras combinando ambos temas.


---
## 1. Diccionarios en Python
Un diccionario permite almacenar datos en pares de **`clave: valor`**. A diferencia de las listas (que usan índices numéricos `0, 1, 2...`), los diccionarios utilizan las **claves** (nombres) para acceder rápidamente a la información.

### 1.1 Creación y Acceso

```python
# Creando un diccionario usando llaves {}
frutas = {
    'Platano': 1.35,
    'Manzana': 0.80,
    'Pera': 0.85
}

# Accediendo a un elemento (forma directa)
print("El precio de la Manzana es: $", frutas['Manzana'])

# Usando get() (es más seguro, no lanza error si la clave no existe)
print("El precio de la Naranja es: $", frutas.get('Naranja', 'No disponible'))
```

### 1.2 Añadir, Modificar y Eliminar
Los diccionarios son **mutables**, lo que significa que podemos cambiar su contenido después de haberlos creado.

```python
datos_personales = {}

# Añadir elementos
datos_personales['nombre'] = 'Juan'
datos_personales['edad'] = 25

# Modificar un elemento existente
datos_personales['edad'] = 26

# Eliminar un elemento usando pop()
edad_eliminada = datos_personales.pop('edad')

print("Diccionario actual:", datos_personales)
print("Valor eliminado:", edad_eliminada)
```

### 1.3 Recorrer un Diccionario
Podemos iterar (recorrer) las claves, los valores, o ambos a la vez usando `.items()`.

```python
asignaturas = {'Matemáticas': 6, 'Física': 4, 'Química': 5}

print("--- Recorriendo Claves y Valores ---")
for materia, creditos in asignaturas.items():
    print(f"{materia} tiene {creditos} créditos.")

# Calcular el total de créditos usando .values() y sum()
total_creditos = sum(asignaturas.values())
print(f"\nTotal de créditos: {total_creditos}")
```

### 1.4 Diccionarios Anidados
Podemos guardar un diccionario dentro de otro diccionario. ¡Es ideal para estructurar datos complejos!

```python
estudiantes = {
    'Juan': {'edad': 20, 'carrera': 'Informática', 'nota': 8.5},
    'María': {'edad': 21, 'carrera': 'Matemáticas', 'nota': 9.2}
}

print("Nota de Juan:", estudiantes['Juan']['nota'])
print("Carrera de María:", estudiantes['María']['carrera'])
```

## 2. Manejo de Excepciones
Los usuarios pueden ingresar datos incorrectos o pueden ocurrir imprevistos. Si no los manejamos, **nuestro programa se caerá**.

Usamos la estructura `try` y `except` para capturar estos errores de forma elegante.

### 2.1 Errores comunes: `ValueError`, `ZeroDivisionError`

```python
try:
    # Intentamos hacer algo que podría fallar
    numero = int(input("Ingresa un número para dividir a 10: "))
    resultado = 10 / numero
    print(f"El resultado es: {resultado}")

except ValueError:
    # Si ingresa letras en lugar de números
    print("❌ Error: Debes ingresar un valor numérico.")

except ZeroDivisionError:
    # Si ingresa 0
    print("❌ Error: ¡No se puede dividir entre cero!")
```

### 2.2 Bloques `else` y `finally`
* `else`: Se ejecuta **SÓLO** si no hubo ningún error.
* `finally`: Se ejecuta **SIEMPRE**, pase lo que pase. Sirve para limpiar recursos (cerrar archivos, etc).

```python
def dividir_seguro(a, b):
    try:
        res = a / b
    except ZeroDivisionError:
        print("Cuidado: División por cero evitada.")
    else:
        print(f"¡Éxito! El resultado es {res}")
    finally:
        print("Operación finalizada.\n")

dividir_seguro(10, 2)
dividir_seguro(10, 0)
```

### 2.3 Provocar errores a propósito: `raise`
A veces queremos validar reglas de nuestro negocio (por ejemplo, "la edad no puede ser negativa") y rechazar ese dato manualmente usando la palabra `raise`.

```python
def validar_edad(edad):
    if edad < 0:
        # Lanzamos nuestro propio error
        raise ValueError(f"La edad no puede ser negativa (Ingresaste: {edad})")
    return edad

try:
    mi_edad = validar_edad(-4)
except ValueError as e:
    print(f"Validación fallida: {e}")
```

## 3. PROYECTO INTEGRADOR: Registro de Compras

Este programa registrará compras (guardando Diccionarios dentro de una Lista) y validará que los montos ingresados sean correctos (usando Excepciones).

```python
def validar_monto(monto_str):
    """Valida que el monto sea un número válido y mayor a cero."""
    try:
        monto_float = float(monto_str)
    except ValueError:
        raise TypeError("El monto debe ser un número.")

    if monto_float <= 0:
        raise ValueError("El monto debe ser mayor a 0.")

    return monto_float

def validar_rubro(rubro_str):
    """Valida que el rubro esté en la lista permitida"""
    rubro_limpio = rubro_str.strip().lower()
    if rubro_limpio not in ['alimentos', 'bebidas', 'limpieza']:
        raise ValueError("Rubros válidos: alimentos, bebidas, limpieza.")
    return rubro_limpio

def ejecutar_sistema():
    compras = [] # Lista que contendrá diccionarios

    print("=== SISTEMA DE REGISTRO DE COMPRAS ===")
    print("(Escribe 'salir' en el rubro para terminar)\n")

    while True:
        # 1. Solicitar y Validar Rubro
        rubro = input("Rubro (Alimentos/Bebidas/Limpieza): ")
        if rubro.lower() == 'salir':
            break

        try:
            rubro_validado = validar_rubro(rubro)
        except ValueError as e:
            print(f"\u25a0 Error: {e}\n")
            continue # Vuelve al inicio del while

        # 2. Solicitar y Validar Monto
        monto = input("Monto de la compra ($): ")
        try:
            monto_validado = validar_monto(monto)
        except (ValueError, TypeError) as e:
            print(f"\u25a0 Error: {e}\n")
            continue

        # 3. Guardar en diccionario
        compra = {
            'rubro': rubro_validado,
            'monto': monto_validado
        }
        compras.append(compra)
        print("\u2705 Compra registrada con éxito.\n")

    # --- ANÁLISIS FINAL (Uso de listas y diccionarios) ---
    print("\n=== RESUMEN FINAL ===")

    # Sumar totales usando comprensión
    total_general = sum(c['monto'] for c in compras)
    print(f"Total ingresado: ${total_general:.2f}")

    # Contar compras por rubro usando diccionarios
    conteo_rubros = {'alimentos': 0, 'bebidas': 0, 'limpieza': 0}
    for c in compras:
        conteo_rubros[c['rubro']] += 1

    for rubro, cant in conteo_rubros.items():
        print(f" - {rubro.capitalize()}: {cant} compra(s)")

# ¡Quita el comentario a la línea de abajo para probar el sistema!
ejecutar_sistema()
```
