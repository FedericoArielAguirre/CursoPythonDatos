# 🚀 Programación Orientada a Objetos (POO) en Python

**Programación Orientada a Objetos**, un paradigma donde los datos y las operaciones se agrupan en unidades lógicas llamadas **objetos**.

Básicamente, la POO nos permite modelar el mundo real en nuestro código.

---

## 1. ¿Qué es un Objeto?
Un objeto es una unidad de código que agrupa información y acciones. Todo objeto del mundo real (y del software) tiene dos componentes:

1. **Atributos (Características):** Los datos que lo describen (la parte *estática*). Ej: color, marca, saldo.
2. **Métodos (Comportamiento):** Las funciones u operaciones que puede realizar (la parte *dinámica*). Ej: acelerar, frenar, pagar.

¡En Python, casi todo es un objeto! Incluso los tipos de datos que ya conoces.

```python
# Las cadenas (str) y las listas (list) son objetos con sus propios métodos
texto = 'Python'
print(texto.upper())  # upper() es un método del objeto cadena

lista = [1, 2, 3]
lista.append(4)       # append() es un método del objeto lista
print(lista)

# Podemos usar hasattr() para verificar si un objeto tiene un método específico
print("¿El texto tiene el método 'upper'?", hasattr(texto, 'upper'))
print("¿El texto tiene el método 'volar'?", hasattr(texto, 'volar'))

# Nota: Puedes usar dir(texto) para ver TODOS los atributos y métodos de un objeto.
```

## 2. Clases: El molde de los objetos
Una **clase** es una plantilla o molde genérico a partir del cual podemos crear muchos objetos con las mismas características. Al proceso de crear un objeto a partir de una clase se le llama **instanciación**.

Para definir una clase usamos la palabra reservada `class`.

```python
# Definición de una clase sencilla
class Saludo:
    mensaje = "Bienvenido "          # Atributo de la clase

    # Nota: El primer parámetro de los métodos siempre es 'self'
    def saludar(self, nombre):       # Método de la clase
        print(self.mensaje + nombre)

# Instanciación: creamos un objeto 's' a partir del molde 'Saludo'
s = Saludo()
s.saludar('Alf')

print("Tipo de objeto:", type(s))
```

## 3. El Inicializador `__init__` y el parámetro `self`
El método `__init__` es un constructor especial. Se ejecuta **automáticamente** cada vez que se crea un objeto nuevo. Nos sirve para darle los valores iniciales (estado) a nuestro objeto.

El parámetro `self` hace referencia al objeto mismo que está ejecutando la acción.

```python
class Tarjeta:
    def __init__(self, numero, cantidad=0):  # Inicializador
        self.numero = numero    # Creamos el atributo 'numero'
        self.saldo = cantidad   # Creamos el atributo 'saldo'

    def mostrar_saldo(self):
        print(f'La tarjeta {self.numero} tiene un saldo de {self.saldo} €')

# Al crear los objetos, __init__ se llama automáticamente con los argumentos dados
t1 = Tarjeta('1111-2222', 1000)
t2 = Tarjeta('9999-8888')       # Usa el saldo por defecto (0)

t1.mostrar_saldo()
t2.mostrar_saldo()

# También podemos acceder a los atributos directamente:
print("Acceso directo al saldo de t1:", t1.saldo)
```

## 4. Atributos de Clase vs Atributos de Instancia
- **De Instancia:** Se definen *dentro* del `__init__` usando `self.`. Cada objeto tiene los suyos propios (ej. cada círculo tiene un radio diferente).
- **De Clase:** Se definen *fuera* de los métodos. Son valores constantes compartidos por todos los objetos de esa clase (ej. el valor de Pi).

```python
class Circulo:
    pi = 3.14159  # Atributo de clase (igual para todos los círculos)

    def __init__(self, radio):
        self.radio = radio  # Atributo de instancia (distinto para cada círculo)

    def area(self):
        return Circulo.pi * (self.radio ** 2)

c1 = Circulo(2)
c2 = Circulo(3)

print(f"Área del círculo 1: {c1.area()}")
print(f"Área del círculo 2: {c2.area()}")
print(f"Ambos comparten el mismo valor de Pi: {c1.pi} y {c2.pi}")
```

## 5. Representación en texto: `__str__`
El método mágico `__str__` define cómo se mostrará tu objeto cuando uses la función `print()`. Si no lo defines, Python imprimirá un mensaje ilegible con la posición de memoria del objeto.

```python
class TarjetaAmigable:
    def __init__(self, numero, saldo=0):
        self.numero = numero
        self.saldo = saldo

    def __str__(self):
        # Esto se llama automáticamente al usar print(objeto)
        return f'[Tarjeta N° {self.numero} | Saldo: {self.saldo}€]'

mi_tarjeta = TarjetaAmigable('0123456789', 500)

# Imprime la cadena formateada gracias a __str__
print(mi_tarjeta)
```

## 6. Herencia
La herencia permite crear una nueva clase (clase hija) a partir de una existente (clase madre). La hija hereda **todos** los métodos y atributos de su madre, evitando que dupliquemos código, y además puede agregar los suyos propios.

```python
class Tarjeta_Descuento(TarjetaAmigable): # Hereda poniendo la clase madre entre paréntesis
    def __init__(self, numero, descuento, saldo=0):
        self.numero = numero
        self.descuento = descuento
        self.saldo = saldo

    def mostrar_descuento(self): # Método exclusivo de esta clase hija
        print(f'¡Tienes un descuento del {self.descuento}% en todas tus compras!')

t_desc = Tarjeta_Descuento('5555-4444', 15, 2000)

# Usamos un método heredado (el __str__ de TarjetaAmigable)
print(t_desc)

# Usamos un método propio de esta clase
t_desc.mostrar_descuento()

# Podemos usar isinstance() para verificar si un objeto es de cierta clase
print("\n¿t_desc es una Tarjeta_Descuento?", isinstance(t_desc, Tarjeta_Descuento))
print("¿t_desc es una TarjetaAmigable?", isinstance(t_desc, TarjetaAmigable)) # ¡También es True por la herencia!
```

## 7. Polimorfismo (Sobrecarga de Métodos)
El polimorfismo ocurre cuando una clase hija **redefine** (sobreescribe) un método de la clase madre para que se comporte de forma distinta. Así, el mismo nombre de método hace cosas diferentes dependiendo de qué objeto lo ejecute.

```python
class TarjetaNormal:
    def __init__(self, saldo):
        self.saldo = saldo

    def pagar(self, monto):
        self.saldo -= monto
        print(f"Pago normal de ${monto}. Saldo restante: ${self.saldo}")

class Tarjeta_Oro(TarjetaNormal):
    def __init__(self, saldo, descuento):
        self.saldo = saldo
        self.descuento = descuento

    # REDEFINIMOS EL MÉTODO PAGAR (Polimorfismo)
    def pagar(self, monto):
        monto_final = monto * (1 - self.descuento / 100)
        self.saldo -= monto_final
        print(f"Pago Oro de ${monto_final} (aplicando descuento). Saldo restante: ${self.saldo}")

t_normal = TarjetaNormal(1000)
t_oro = Tarjeta_Oro(1000, descuento=10) # 10% de descuento

# Llamamos al mismo método, pero el comportamiento es distinto
t_normal.pagar(100)
t_oro.pagar(100)
```

## 8. Los 4 Pilares de la Programación Orientada a Objetos (Resumen)

Estos son los conceptos universales que sostienen la POO en cualquier lenguaje:

1. **Encapsulación:** Agrupar datos (atributos) y métodos en unidades lógicas (objetos) para evitar manipular los datos de forma descontrolada desde el exterior.
2. **Abstracción:** Ocultar al usuario los detalles complejos de implementación. Solo necesitas saber *qué* hace el método y cómo llamarlo, no *cómo* funciona por dentro.
3. **Herencia:** Crear nuevas clases a partir de clases existentes para evitar la duplicación de código.
4. **Polimorfismo:** La capacidad de que el mismo método actúe de manera diferente dependiendo del objeto sobre el cual se invoque (sobrecarga).

## 9. Código Final Integrador 🏆
A continuación, presentamos un script completo que unifica **todos** los conceptos vistos en este cuaderno.

Imagina un sistema bancario con distintos tipos de cuentas. Analiza cómo la clase `CuentaAhorro` y `CuentaPremium` heredan de `CuentaBancaria`, pero cada una aplica el **polimorfismo** para reaccionar de forma distinta al intentar retirar dinero.

```python
# 1. CLASE MADRE (Molde base)
class CuentaBancaria:
    banco = "PythonBank"  # Atributo de clase (compartido por todas las cuentas)

    def __init__(self, titular, saldo=0):  # __init__ (Inicializador)
        self.titular = titular             # Atributo de instancia
        self.saldo = saldo                 # Atributo de instancia

    def __str__(self):  # Representación en texto (Impresión amigable)
        return f"🏦 {self.banco} | Titular: {self.titular} | Saldo: ${self.saldo}"

    def depositar(self, monto):
        self.saldo += monto
        print(f"[+] {self.titular} depositó ${monto}. Nuevo saldo: ${self.saldo}")

    def retirar(self, monto):
        self.saldo -= monto
        print(f"[-] {self.titular} retiró ${monto}. Nuevo saldo: ${self.saldo}")


# 2. HERENCIA: Clase Hija 1
class CuentaAhorro(CuentaBancaria):
    # POLIMORFISMO: Redefinimos retirar() para que no permita saldo negativo
    def retirar(self, monto):
        if monto > self.saldo:
            print(f"[❌] {self.titular} intentó retirar ${monto}. ¡Fondos insuficientes!")
        else:
            self.saldo -= monto
            print(f"[-] {self.titular} retiró ${monto} de su cuenta de Ahorro. Nuevo saldo: ${self.saldo}")


# 3. HERENCIA: Clase Hija 2
class CuentaPremium(CuentaBancaria):
    def __init__(self, titular, saldo=0, limite_descubierto=500):
        # Usamos super() para llamar al __init__ de la clase madre y no repetir código
        super().__init__(titular, saldo)
        self.limite_descubierto = limite_descubierto

    # POLIMORFISMO: Redefinimos retirar() para permitir saldo negativo hasta el límite
    def retirar(self, monto):
        if (self.saldo - monto) < -self.limite_descubierto:
            print(f"[❌] {self.titular} intentó retirar ${monto}. ¡Límite de descubierto de ${self.limite_descubierto} excedido!")
        else:
            self.saldo -= monto
            print(f"[-] {self.titular} retiró ${monto} de su cuenta Premium. Nuevo saldo: ${self.saldo}")


# --- PRUEBA DEL CÓDIGO (INSTANCIACIÓN Y USO) ---
print("--- CREACIÓN DE OBJETOS ---")
cuenta_ana = CuentaAhorro("Ana", 100)
cuenta_carlos = CuentaPremium("Carlos", 100, limite_descubierto=200)

# Uso del método __str__
print(cuenta_ana)
print(cuenta_carlos)

print("\n--- OPERACIONES (POLIMORFISMO EN ACCIÓN) ---")
cuenta_ana.depositar(50)
cuenta_carlos.depositar(50)

# Ana intenta retirar 200 pero solo tiene 150 (Falla por ser CuentaAhorro)
cuenta_ana.retirar(200)

# Carlos intenta retirar 200 teniendo 150 (Pasa porque es CuentaPremium y tiene límite de descubierto)
cuenta_carlos.retirar(200)
```
